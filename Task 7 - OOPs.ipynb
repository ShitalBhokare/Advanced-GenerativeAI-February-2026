***Problem 1: Smart Light Controller***

**Problem Statement**
Create a class to control a smart light that can be turned ON or OFF and display its status.

**üåç Real-Time Use**
Smart home automation systems

**üì• Sample Input**
Light Name: Bedroom Light
Action: ON

**üì§ Expected Output**
Bedroom Light is ON


```python
class SmartLightController:
    def __init__(self, name):
        self.name = name 
        self.current_status = "OFF"  # Light is off initially

    def switch_on(self):
        self.current_status = "ON"

    def switch_off(self):
        self.current_status = "OFF"

    def show_status(self):
        print(self.name + " is currently " + self.current_status)

#Taking input from user 
light_name = input("Enter light name : ")
action = input("Enter Action (ON/OFF): ")

my_light = SmartLightController(light_name)

if action.upper() == "ON":
    my_light.switch_on()
else:
    my_light.switch_off()
my_light.show_status()
```

***Problem 2: Employee ID Card System***

**Problem Statement**
Create a class that stores employee name, ID, and department and displays ID card details.

**Real-Time Use**
Corporate HR systems

**üì• Sample Input**
Employee Name: Rahul
Employee ID: EMP102
Department: IT

**üì§ Expected Output**
Employee ID Card
Name: Rahul
ID: EMP102
Department: IT



```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name 
        self.emp_id = emp_id
        self.dept = dept

    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name : ", self.name)
        print("ID: ", self.emp_id)
        print("Department: ", self.dept)

#Taking input from user
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
dept = input("Enter Department: ")

#Creating object 
emp = Employee(name, emp_id, dept)

#Displaying ID card
emp.display_id_card()
```

***Problem 3: Mobile Contact Record***

**Problem Statement**
Create a class to store contact name and phone number and display contact information.

**Real-Time Use**
Mobile contact applications

**üì• Sample Input**
Contact Name: Anita
Phone Number: 9876543210

**üì§ Expected Output**
Contact Saved
Name: Anita
Phone: 9876543210


```python
class MobileContact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print("Name: ", self.name)
        print("Phone: ", self.phone)

#Taking input from user
contact_name = input("Enter Name: ")
phone_number = input("Enter Phone number : ")

# Creating object
contact = MobileContact(contact_name, phone_number)

# Displaying contact details
contact.display_contact()
```

***Problem 4: Product Price Tag Generator***

**Problem Statement**
Create a class that stores product name and price and prints a formatted price tag.

**Real-Time Use**
Retail billing systems

**üì• Sample Input**
Product Name: Headphones
Price: 2499

**üì§ Expected Output**
Product: Headphones
Price: ‚Çπ2499


```python
class ProductPriceGenerator:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def display_product_price(self):
        print("\nProduct: ", self.product_name)
        print("Price: ", self.price)

# Taking input
p_name = input("Enter Product Name : ")
p_price = int(input("Enter price of the product: "))

#Creating object
product_price = ProductPriceGenerator(p_name, p_price)

# Displaying formatted price tag
product_price.display_product_price()
    
```

***Problem 5: Movie Rating Display System***

**Problem Statement**
Create a class that stores movie name and rating and displays movie details.

**Real-Time Use**
Streaming platforms

**üì• Sample Input**
Movie Name: Inception
Rating: 4.8

**üì§ Expected Output**
Movie: Inception
Rating: 4.8 / 5


```python
class MovieRatingSystem:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display_movie_details(self):
        print("\nMovie: ",self.movie_name)
        print(f"Rating: {self.rating} / 5")

# Taking input from user
name = input("Enter movie name : ")
rating = float(input("Enter Rating: "))

# Creating object
movie = MovieRatingSystem(name, rating)

# Displaying movie details
movie.display_movie_details()
```

***Problem 6: Delivery Address Manager***

**Problem Statement**
Create a class that stores customer name and delivery address and prints delivery details.

**Real-Time Use**
Courier and logistics platforms

**üì• Sample Input**
Customer Name: Suman
Address: Hyderabad

**üì§ Expected Output**
Delivery Details
Customer: Suman
Address: Hyderabad


```python
class DeliveryAddressManager:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("\nDelivery Details")
        print("Cutomer: ", self.customer_name)
        print("Address: ", self.address)

# Taking input from user
name = input("Enter the customer name: ")
address = input("Enter customer address: ")

# Creating object
customer = DeliveryAddressManager(name, address)

# Displaying delivery details
customer.display_delivery_details()
```
