***Problem 1: Unique Words in a Sentence***

**Problem Statement**
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves
Hint
Split the sentence into individual words.
Convert the list of words into a set to remove duplicates.
Print both the set and its length.

**Input**
Enter sentence: python is easy and python is powerful

**Expected Output**
Unique words count: 5  
Unique words: {'python', 'is', 'easy', 'and', 'powerful'}


```python
sentence = "python is easy and python is powerful"

#split() converts the sentence into words
words = sentence.split()

#empty list where unique words will store 
unique_words = []

for word in words:
    if word not in unique_words:            # This prevents repetition of words
        unique_words.append(word)           # Add unique words to the list

unique_words_count = len(unique_words)      # Total number of unique words 

unique_words = "{" + ", ".join(f"'{word}'" for word in unique_words) + "}" 

# Display results
print(f"Unique words count : {unique_words_count}")  
print(f"Unique words : {unique_words}")
```

***Problem 2: Highest Salary from Employee Data***

**Problem Statement**
Store employee names and their salaries, and determine which employee has the highest salary.
Hint
Use a dictionary where the key is the employee name and the value is the salary.
Compare salary values to find the highest.

**Input Format**
A dictionary where:
Keys represent employee names
Values represent employee salaries

**Example Input**
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

**Expected Output**
Highest Salary: Ravi - 75000



```python
employees = {"Ravi" : 75000, "Anita" : 68000, "Kiran" : 72000}

#variable to store the name of the employee with the highest salary
employee_name = ""  

#variable to store the highest salary 
highest_salary = 0

for name, salary in employees.items():
    # check if current salary is greater than highest_salary
    if salary > highest_salary:
        employee_name = name       # update employee name
        highest_salary = salary    # update highest salary

# Display employee with highest salary
print(f"{employee_name} - {highest_salary}")
```

***Problem 3: Find Maximum and Minimum Values***

**Problem Statement**
Find the maximum and minimum values from a list of numbers.
Hint
Initialize both maximum and minimum with the first element of the list.
Use a loop to compare each element.

**Input Format**
A list of integers where each element represents a numeric value.

**Example Input**
numbers = [45, 22, 89, 10, 66]

**Expected Output**
List: [45, 22, 89, 10, 66]  
Max: 89  
Min: 10


```python
numbers = [45, 22, 89, 10, 66]

# Initalizing max_num and min_num with first element
max_num = numbers[0]
min_num = numbers[0]

for num in numbers:
    if num > max_num:   # check if current num is greater than max_num
        max_num = num   # update max_num
    if num < min_num:   # check if current num is smaller than min_num
        min_num = num   # update min_num

# Display result
print(f"List : {numbers}")
print(f"Max : {max_num}" )
print(f"Min : {min_num}" )
```

***Problem 4: Count Products Above a Price Threshold***
**Problem Statement**
Count how many products have a price greater than 1000.
Hint
Use a loop to compare each price with 1000.
**Input Format**
A list of integers where each element represents the price of a product.
**Example Input**
prices = [450, 1200, 899, 1500, 300]

**Expected Output**
Products above 1000: 2


```python
prices = [450, 1200, 899, 1500, 300]

# variable to count products with price greater than 1000
count = 0

for price in prices:
    if price > 1000:         # check if the price is greater than 1000
        count = count + 1    # Increase the count if condition is true

# Display result
print(f"Products above 1000 : {count}")
```

***Problem 5: Calculate Attendance Percentage***

**Problem Statement**
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).
Hint
Count the number of "P" values.
Divide by the total number of days.

**Example Input**
attendance = ["P", "P", "A", "P", "P"]

**Expected Output**
Attendance Percentage: 80.0


```python
attendance = ["P", "P", "A", "P", "P"]
total = len(attendance)     # total items in the list
count = 0                   # Variable to count number of "Present" days 

for atten in attendance:
    if atten == "P":       # check if the student was present
        count += 1         # Increase count for each "P"

# Calculate attendance percentage
atten_percentage = (count/total)*100

# display result
print(f"Attendance Percentage : {atten_percentage}")
```

***Problem 6: Remove Duplicate Phone Numbers***

**Problem Statement**
You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.
Hint
Convert the list into a set.
Sets automatically remove duplicate values.

**Input Format**
A list of integers where each integer represents a phone number.

**Example Input**
phone_numbers = [9876543210, 9123456789, 9876543210]

**Expected Output**
Unique phone numbers: {9876543210, 9123456789}


```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = []                   # Empty list to store unique numbers 

for number in phone_numbers:
    if number not in unique_numbers:  # Check if number is already in the list
        unique_numbers.append(number) # Add number if it is not a duplicate

unique_phone_numbers = "{" + ", ".join (str(number) for number in unique_numbers) +"}"

# Display result
print(f"Unique phone numbers : {unique_phone_numbers}")
```

***Problem 7: Count Character Frequency***

**Problem Statement**
Count how many times each character appears in a given string.
Hint
Use a dictionary.
Increment the count for each character occurrence.

**Expected Output**
{'p': 2, 'y': 1, 't': 1, 'h': 1, 'o': 1, 'n': 1}


```python
word = "ppython"

count = {}  # Empty dictionary to store character frequency

for ch in word :
    if ch in count:      # Check if character already exists in dictionary
        count[ch] += 1   # Increase its count
    else:
        count[ch] = 1    # Add charater to dictionary with initial count 1

# Display result
print(count)
```

***Problem 10: Calculate Average Marks***

**Problem Statement**
Calculate the average marks of a student from a list of marks.
Hint
Calculate the sum of all values.
Divide by the total number of marks.

**Expected Output**
Average Marks: 72.5



```python
marks = [65, 70, 75, 80]
count = 0                 # Variable to store sum
total = len(marks)        # Total number of marks

for mark in marks:
    count += mark        # Add each mark to count

avg = (count / total)    # Calculate average

# Display result
print(f"Average marks : {avg}")
```
