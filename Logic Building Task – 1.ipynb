**1.User Login Check**

Problem Statement: Given a username and password, check whether login is successful.

username = "admin"
password = "1234"

Requirements: Print "Login Successful" if both username and password match Otherwise print "Invalid Credentials"

```python
username = "admin"
password = "1234"

# Taking input from user
uname = input("Enter Username: ")
pword = input("Enter Password: ")

# Check if entered credentials match the predefined one
if(uname == username and pword == password):
    print("Login Successfully")
else:
    print("Invalid Credentials")
```

**2. Pass / Fail Analyzer**

Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks â‰¥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly

Real-World Application: Academic evaluation systems


```python
marks = [45, 78, 90, 33, 60]

passed_students = 0
failed_students = 0

# Iterate through each student's marks
for s_marks in marks:
    if s_marks >= 50:   # Check pass condition (marks >= 50)
        passed_students += 1
    else:
        failed_students += 1

# Display final result
print(f"Passed Students : {passed_students}")
print(f"Failed Students: {failed_students}")
         
```

**3. Simple Data Cleaner**

Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list

Real-World Application: Data preprocessing before analysis


```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_name_list = []

for n in names:
    n = n.strip()   #remove spaces
    n = n.lower()   #convert all names to lowercase
    cleaned_name_list.append(n)
    
print(f"Cleaned name list: {cleaned_name_list}")
```

**4. Message Length Analyzer**

Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters

Real-World Application: Text filtering and validation systems


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    print(f"Message: {msg}")
    print(f"Message Length: {len(msg)}")

    # Flag messages longer than 10 characters
    if len(msg) >10 :
        print("Flag : This message is longer than 10 characters. ")

    print()
```

**5. Error Message Detector**
    
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

total_error_entries = 0

# Count how many times "ERROR" appears in logs
for log in logs:
    if log == "ERROR":
        total_error_entries += 1

print(f"Total Error Entries : {total_error_entries}")
```
