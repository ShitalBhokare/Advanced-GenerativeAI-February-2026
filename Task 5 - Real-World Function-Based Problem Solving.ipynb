***Problem 1: Social Media – Post Engagement Analyzer***

**Problem Statement**
Create a function to analyze engagement on a social media post.

**Rules**
Calculate total likes
If Total Likes ≥ 1000 → Viral Post
Otherwise → Normal Engagement

**Real-Time Use**
Social media analytics platforms
Marketing dashboards

**Hint**
Use function
Use loop for calculations

**Expected Output Format**
Total Likes: 1050
Post Status: Viral Post

```python
def analyze_post_engagement():
    # Initialize total likes to 0
    total_likes = 0

    # Ask user how many posts to analyze
    n = int(input("Enter number of posts: "))

    # Loop to collect likes for each post
    for i in range(n):
        # Take likes input for each post
        likes = int(input(f"Enter likes for post {i+1}: "))
        
        # Add likes to total_likes
        total_likes += likes

    # Print total likes
    print("\nTotal Likes:", total_likes)

    # Check if post is viral or normal
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Call the function
analyze_post_engagement()
```

***Problem 3: Agriculture – Rainfall Adequacy Checker***

**Problem Statement**
Create a function that analyzes seasonal rainfall data.

**Rules**
Calculate average rainfall
Average ≥ required level → Adequate Rainfall
Otherwise → Inadequate Rainfall

**Real-Time Use**
Climate monitoring systems
Irrigation planning tools

**Hint**
Use function
Use loop and math operations

**Expected Output Format**
Average Rainfall: 72
Rainfall Status: Adequate Rainfall

```python
def rainfall_checker():
    total_rainfall = 0

    # Number of months in season
    n = int(input("Enter number of months: "))

    # Required rainfall level
    required = float(input("Enter required average rainfall level: "))

    # Collect rainfall data
    for i in range(n):
        rainfall = float(input(f"Enter rainfall for month {i+1}: "))
        total_rainfall += rainfall

    # Calculate average rainfall
    average = total_rainfall / n

    # Print average rainfall
    print("\nAverage Rainfall:", round(average, 2))

    # Check rainfall adequacy
    if average >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Function Call
rainfall_checker()
```

***Problem 4: Social Media – Duplicate Account Detection***

**Problem Statement**
Create a function to detect duplicate usernames in a system.

**Rules**
If duplicates exist → Flag accounts
Else → Accounts are unique

**Real-Time Use**
Platform moderation
Fraud detection systems

**Hint**
Use function
Use set data structure

**Expected Output Format**
Duplicate Accounts Found: Yes


```python
def detect_duplicate_accounts():
    usernames = []

    # Number of users
    n = int(input("Enter number of usernames: "))

    # Collect usernames
    for i in range(n):
        name = input(f"Enter username {i+1}: ")
        usernames.append(name)

    # Convert list to set
    unique_usernames = set(usernames)

    # Check for duplicates
    if len(usernames) != len(unique_usernames):
        print("\nDuplicate Accounts Found: Yes")
    else:
        print("\nDuplicate Accounts Found: No")


# Function Call
detect_duplicate_accounts()
```

***Problem 6: Agriculture – Premium Crop Price Filter***

**Problem Statement**
Create a function that filters crop prices above a premium threshold.

**Rules**
Price > ₹2000 → Premium Crop
Else → Regular Crop

**Real-Time Use**
Market analytics
Export pricing systems

**Hint**
Use function
Use list filtering logic

**Expected Output Format**
Premium Crops: [2500, 3200]


```python
def filter_premium_crops():
    prices = []

    # Number of crops
    n = int(input("Enter number of crop prices: "))

    # Collect crop prices
    for i in range(n):
        price = int(input(f"Enter price for crop {i+1}: "))
        prices.append(price)

    # Filter premium crops (Price > 2000)
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    # Print premium crops
    print("\nPremium Crops:", premium_crops)


# Function Call
filter_premium_crops()
```

***Problem 9: E-Learning – Student Attendance Eligibility System***

**Problem Statement**
Create a function to calculate attendance percentage and check exam eligibility.

**Rules**
Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible

**Real-Time Use**
Learning management systems
Internship tracking platforms

**Hint**
Use function
Use list and loop

**Expected Output Format**
Attendance Percentage: 80.0
Exam Eligibility: Eligible

```python
def check_attendance_eligibility():
    attendance_list = []

    # Number of classes conducted
    total_classes = int(input("Enter total number of classes: "))

    # Collect attendance data (1 = Present, 0 = Absent)
    for i in range(total_classes):
        status = int(input(f"Enter attendance for class {i+1} (1=Present, 0=Absent): "))
        attendance_list.append(status)

    # Calculate total present days
    total_present = 0
    for status in attendance_list:
        total_present += status

    # Calculate attendance percentage
    percentage = (total_present / total_classes) * 100

    # Print attendance percentage
    print("\nAttendance Percentage:", round(percentage, 1))

    # Check eligibility
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Function Call
check_attendance_eligibility()
```

***Problem 10: Smart Electricity Bill Analyzer***

**Problem Statement**
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

**Billing Rules**
First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit

**Usage Classification**
Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage

**Real-Time Use**
Smart meter systems
Electricity board billing software
Energy consumption analytics

**Hint**
Use function to calculate bill
Use conditional logic for slabs
Use return values (bill and usage status)


```python
def calculate_electricity_bill(units):
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# ---- Main Program ----
units = int(input("Enter total units consumed: "))

total_bill, usage_status = calculate_electricity_bill(units)

print("Total Electricity Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
